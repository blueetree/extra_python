```python
import pandas as pd
import numpy as np
import matplotlib as mpl
mpl.use('TkAgg')
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import seaborn as sns
%matplotlib inline
from datetime import datetime
```

```python
events = pd.read_csv('app_events.csv') 
locations = pd.read_csv('app_locations.csv') 
```

# EDA for app_events data

```python
events.head()
```

```python
# check duplication
events[events.duplicated('id')]
```

```python
events.info()
```

```python
# convert time
events["timestamp"] = events["timestamp"].apply(lambda x: datetime.strptime(str(x),"%Y-%m-%d %H:%M:%S"))
type(events["timestamp"][0])
events['year'] = events["timestamp"].apply(lambda x: x.year)
events['month'] = events["timestamp"].apply(lambda x: x.month)
events['quarter'] = pd.PeriodIndex(events['timestamp'], freq='Q')
events['quarter'] = events["quarter"].astype(str).apply(lambda x: x[-2:])
events['weekday'] = events["timestamp"].apply(lambda x: x.strftime('%A'))
events['day'] = events["timestamp"].apply(lambda x: x.day)
events["hour"] = events["timestamp"].apply(lambda x: x.hour)
```

```python
# check outliers
events.describe()
```

```python
# check nan
events.isnull().sum()
```

```python
events[events.isna().any(axis=1)]
```

```python
# remove nan: it has a very small amount of nan(10/16050)
events = events[np.isfinite(events['user_id'])]
```

```python
events.isnull().sum()
```

```python
# convert ids and category to string
events[['id', 'app_id', 'user_id','category_id']] = events[['id', 'app_id', 'user_id','category_id']].astype(int).astype(str)
```

```python
fig = plt.figure()
fig.set_size_inches(10,20)
ax1 = fig.add_subplot(121)
grouped = events[events['event']=='Installed'].groupby('category_id')['app_id'].count()
patches, _, _ = ax1.pie(grouped, autopct='%1.1f%%', pctdistance=1.1, labeldistance=1.2)
ax1.legend(patches, grouped.index, loc=10)
ax1.set_title("Installation Number by Category")
ax2 = fig.add_subplot(122)
grouped = events[events['event']=='Uninstalled'].groupby('category_id')['app_id'].count()
patches, _, _ = ax2.pie(grouped, autopct='%1.1f%%', pctdistance=1.1, labeldistance=1.2)
ax2.legend(patches, grouped.index, loc=10)
ax2.set_title("Uninstallation Number by Category")
plt.show()
```

```python
sns.countplot(x='category_id', hue='event', data=events).set_title("Installation Status by Category")
```

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))
sns.countplot(x='year', hue='event', data=events, ax=ax[0,0]).set_title("Installation Status by Year")
sns.countplot(x='month', hue='event', data=events, ax=ax[0,1]).set_title("Installation Status by Month")
sns.countplot(x='quarter', hue='event', data=events, ax=ax[1,0]).set_title("Installation Status by Quarter")
sns.countplot(x='weekday', hue='event', data=events, ax=ax[1,1]).set_title("Installation Status by Day of Week")
sns.countplot(x='day', hue='event', data=events, ax=ax[2,0]).set_title("Installation Status by Day of Day")
sns.countplot(x='hour', hue='event', data=events, ax=ax[2,1]).set_title("Installation Status by Hour")
plt.tight_layout()
plt.show()
```

```python
sns.set(rc={'figure.figsize':(18,8)})
sns.countplot(x='state_abbrev', data=locations).set_title("Market Share by State")
```
